The file IMDB_Spanish_1500_Balanced.csv contains 1,500 movie reviews with sentiment annotations. It has 5 columns:

* Unnamed: 0 – An index/reference column from the original source (review IDs).
* review_en – The original movie review text in English.
* review_es – The Spanish translation of the same review.
* sentiment – The sentiment label in English (positive / negative).
* sentimiento – The sentiment label in Spanish (positivo / negativo).
* Dataset Balance 900 positive reviews (positivo)
* 600 negative reviews (negativo)

This makes it a balanced sentiment dataset suitable for text classification and multilingual NLP tasks, such as machine translation quality, sentiment analysis in Spanish/English, and transfer learning.

# Data Cleaning and Normalization for English and Spanish Reviews

```python
import pandas as pd

url = "https://raw.githubusercontent.com/Eman-Nasr/multilingual-movie-reviews/main/multilingual_reviews_cleaned.csv"
df = pd.read_csv(url)
df.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download NLTK data if not available
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize tools
stop_words_en = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_english_text(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # remove punctuation & symbols
    tokens = [word for word in text.split() if word not in stop_words_en]  # remove stopwords
    tokens = [lemmatizer.lemmatize(word) for word in tokens]  # lemmatize
    return tokens

# Apply cleaning function
df['review_en_tokens'] = df['review_en'].apply(clean_english_text)

# Display result
df[['review_en', 'review_en_tokens']].head()
```

```python
import spacy
nlp_es = spacy.load("es_core_news_sm")

def clean_spanish_text(text):
    text = text.lower()
    text = re.sub(r'[^a-záéíóúñü\s]', '', text)
    doc = nlp_es(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop]
    return tokens

df['review_es_tokens'] = df['review_es'].apply(clean_spanish_text)
df[['review_es', 'review_es_tokens']].head()
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/Eman-Nasr/multilingual-movie-reviews/main/multilingual_reviews_cleaned.csv"
df = pd.read_csv(url)

print(df.columns.tolist())
df.head()
```

```python
import re
import string
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

stop_en = set(stopwords.words('english'))
stop_es = set(stopwords.words('spanish'))

def clean_text(text, lang='en'):
    text = str(text).lower()
    text = re.sub(r'http\S+', '', text)                 # remove URLs
    text = re.sub(f"[{re.escape(string.punctuation)}]", '', text)  # remove punctuation
    text = re.sub(r'\d+', '', text)                     # remove numbers
    words = [w for w in text.split() if w not in (stop_en if lang == 'en' else stop_es)]
    return ' '.join(words)

# Apply cleaning
df['review_en_clean'] = df['review_en'].apply(lambda x: clean_text(x, 'en'))
df['review_es_clean'] = df['review_es'].apply(lambda x: clean_text(x, 'es'))

df[['review_en', 'review_en_clean', 'review_es', 'review_es_clean']].head(3)
```

```python
df['review_en_wordcount'] = df['review_en_clean'].apply(lambda x: len(str(x).split()))
df['review_es_wordcount'] = df['review_es_clean'].apply(lambda x: len(str(x).split()))

avg_en = df['review_en_wordcount'].mean()
avg_es = df['review_es_wordcount'].mean()

print("Average English word count:", round(avg_en, 2))
print("Average Spanish word count:", round(avg_es, 2))
```

```python
# Compares the average review length between English and Spanish datasets to highlight language verbosity differences.

import matplotlib.pyplot as plt

plt.bar(['English', 'Spanish'], [avg_en, avg_es], color=['skyblue', 'salmon'])
plt.title("Average Word Count per Review (English vs Spanish)")
plt.ylabel("Average Word Count")
plt.show()
```

```python
# Visualizes the balance of positive and negative reviews in both English and Spanish datasets to check for class imbalance.

import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# English sentiment plot
sns.countplot(x='sentiment', data=df, palette='pastel', ax=axes[0])
axes[0].set_title('Sentiment Distribution (English Reviews)')
axes[0].set_xlabel('Sentiment')
axes[0].set_ylabel('Count')

# Spanish sentiment plot
sns.countplot(x='sentimiento', data=df, palette='pastel', ax=axes[1])
axes[1].set_title('Sentiment Distribution (Spanish Reviews)')
axes[1].set_xlabel('Sentiment')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Displays how review lengths are distributed for each language, helping detect typical review size and extreme values.

import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# English word count distribution
sns.histplot(df['review_en_wordcount'], bins=40, kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Distribution of Word Count per Review (English)')
axes[0].set_xlabel('Number of Words')
axes[0].set_ylabel('Number of Reviews')

# Spanish word count distribution
sns.histplot(df['review_es_wordcount'], bins=40, kde=True, color='salmon', ax=axes[1])
axes[1].set_title('Distribution of Word Count per Review (Spanish)')
axes[1].set_xlabel('Number of Words')
axes[1].set_ylabel('Number of Reviews')

plt.tight_layout()
plt.show()
```

```python
# Shows the ten most common words in English and Spanish reviews after cleaning, revealing dominant vocabulary and patterns.

from collections import Counter
import seaborn as sns

def top_words(texts, n=10):
    all_words = ' '.join(texts).split()
    return Counter(all_words).most_common(n)

top_en = pd.DataFrame(top_words(df['review_en_clean']), columns=['Word', 'Frequency'])
top_es = pd.DataFrame(top_words(df['review_es_clean']), columns=['Word', 'Frequency'])

top_en['Language'] = 'English'
top_es['Language'] = 'Spanish'
combined = pd.concat([top_en, top_es])

sns.barplot(x='Frequency', y='Word', hue='Language', data=combined)
plt.title("Top 10 Frequent Words (English vs Spanish)")
plt.show()
```

```python
# Compares how review length varies by sentiment for each language, showing if longer texts correlate with certain emotions.

import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# English reviews
sns.boxplot(
    x='sentiment',
    y='review_en_wordcount',
    data=df,
    palette='coolwarm',
    ax=axes[0]
)
axes[0].set_title('Review Length by Sentiment (English)')
axes[0].set_xlabel('Sentiment')
axes[0].set_ylabel('Word Count')

# Spanish reviews
sns.boxplot(
    x='sentimiento',
    y='review_es_wordcount',
    data=df,
    palette='coolwarm',
    ax=axes[1]
)
axes[1].set_title('Review Length by Sentiment (Spanish)')
axes[1].set_xlabel('Sentiment')
axes[1].set_ylabel('Word Count')

plt.tight_layout()
plt.show()
```

```python
# Summarizes the general spread of review lengths for both languages, providing an overview of writing patterns.

import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

sns.boxplot(y='review_en_wordcount', data=df, color='lightgreen', ax=axes[0])
axes[0].set_title('Review Length – All English Reviews')
axes[0].set_ylabel('Word Count')

sns.boxplot(y='review_es_wordcount', data=df, color='lightgreen', ax=axes[1])
axes[1].set_title('Review Length – All Spanish Reviews')
axes[1].set_ylabel('Word Count')

plt.tight_layout()
plt.show()
```

```python
# Removes extremely long reviews to focus on the typical range of review lengths, improving clarity of sentiment comparisons.

df_trim_en = df[df['review_en_wordcount'] < 600]
df_trim_es = df[df['review_es_wordcount'] < 600]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.boxplot(x='sentiment', y='review_en_wordcount', data=df_trim_en, palette='coolwarm', ax=axes[0])
axes[0].set_title('Review Length by Sentiment (English, Trimmed Outliers)')
axes[0].set_xlabel('Sentiment')
axes[0].set_ylabel('Word Count')

sns.boxplot(x='sentimiento', y='review_es_wordcount', data=df_trim_es, palette='coolwarm', ax=axes[1])
axes[1].set_title('Review Length by Sentiment (Spanish, Trimmed Outliers)')
axes[1].set_xlabel('Sentiment')
axes[1].set_ylabel('Word Count')

plt.tight_layout()
plt.show()
```
